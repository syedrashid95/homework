```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
merged_df = pd.merge(ride_data, city_data, on = "city", how = "left")
# Display the data table for preview
```

```python
merged_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_city = merged_df[merged_df["type"] == "Urban"].groupby([merged_df["city"]])
rural_city = merged_df[merged_df["type"] == "Rural"].groupby([merged_df["city"]])
suburban_city = merged_df[merged_df["type"] == "Suburban"].groupby([merged_df["city"]])
# Build the scatter plots for each city types
x_urban = urban_city["ride_id"].count()
y_urban = urban_city["fare"].mean()
s_urban = urban_city["driver_count"].mean()
x_rural = rural_city["ride_id"].count()
y_rural = rural_city["fare"].mean()
s_rural = rural_city["driver_count"].mean()
x_suburban = suburban_city["ride_id"].count()
y_suburban = suburban_city["fare"].mean()
s_suburban = suburban_city["driver_count"].mean()
plt.scatter(x_urban, y_urban, label = "Urban", s=s_urban * 10, color=["coral"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(x_rural, y_rural, label = "Rural", s=s_rural * 10, color=["gold"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(x_suburban, y_suburban, label = "Suburban", s=s_suburban  * 10,  color=["lightskyblue"], edgecolor="black", alpha = 0.75, marker="o")
plt.grid()
# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
# Create a legend
legend = plt.legend(fontsize = 8, title= "City Types", loc="upper right") 
# Incorporate a text label regarding circle size
plt.text(43,35,"Note: \nCircle size correlates with driver count per city.", fontsize = 10)
# Saved Figure
plt.savefig("../Images/pyber_starter4.png")
```

## Total Fares by City Type

```python
# Calculate Type Percents
city_type = merged_df.groupby(["type"])
fare_total = city_type["fare"].sum()
# Build Pie Chart

types = ["Rural", "Suburban", "Urban"]
explode = (0, 0, 0.1)
colors = ["gold", "lightskyblue", "lightcoral"]
plt.pie(fare_total, explode=explode, labels= types, colors = colors, 
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total fares by City Type")
# Save Figure
plt.savefig("../Images/pyber_starter.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_total = city_type['ride_id'].count()
# Build Pie Chart
types = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.2)
colors = ["gold", "lightskyblue", "lightcoral"]
plt.pie(rides_total, explode=explode, labels= types, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("../Images/pyber_starter1.png")
```
